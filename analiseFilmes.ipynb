```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#importando base de dados

df = pd.read_csv("./csv/dadosNetflix.csv")
```

```python
#excluindo coluna language e Age
df = df.drop(columns=['Language'])
df = df.drop(columns=['Age'])
df
```

```python
#removendo linhas com algum valor null 

df = df.dropna(how= 'any', axis = 0)

print("dataframe: ", df.shape)
df.info()
```

```python
#filtrar: eua, india, brasil e nigeria

df_eua = df.loc[(df['Country'] == 'United States')]

df_india = df.loc[(df['Country'] == 'India')]

df_brasil = df.loc[(df['Country'] == 'Brazil')]

df_nigeria = df.loc[(df['Country'] == 'Nigeria')]
```

```python
plt.figure(figsize=(8, 6)) 
years = df.Year.unique()

classif_brasil = df_brasil.groupby('Year')['IMDb'].mean()
classif_eua = df_eua.groupby('Year')['IMDb'].mean()
classif_india = df_india.groupby('Year')['IMDb'].mean()
classif_nigeria = df_nigeria.groupby('Year')['IMDb'].mean()

ax = sns.lineplot(x=years, y=classif_brasil, label='Brasil')
ax = sns.lineplot(x=years, y=classif_eua, label='Estados Unidos')
ax = sns.lineplot(x=years, y=classif_india, label='India')
ax = sns.lineplot(x=years, y=classif_nigeria, label='Nigeria')

ax = sns.scatterplot(x=years, y=classif_brasil, color='blue')
ax = sns.scatterplot(x=years, y=classif_eua, color='orange')
ax = sns.scatterplot(x=years, y=classif_india, color='green')
ax = sns.scatterplot(x=years, y=classif_nigeria, color='red')

plt.legend(title='Pa√≠ses', prop={'size': 14},  title_fontsize = 16)
plt.ylabel("IMDb", size=19)
plt.xlabel("Anos", size=19)
plt.xticks([2018, 2019, 2020, 2021], size=16)
plt.yticks([5.5, 6.0, 6.5, 7.0, 7.5], size=16)
plt.show()
```

```python
plt.figure(figsize=(8, 6)) 
eua = df_eua.Country.unique()
india = df_india.Country.unique()
nigeria = df_nigeria.Country.unique()
brasil = df_brasil.Country.unique()

classif_brasil = df_brasil['IMDb'].mean()
classif_eua = df_eua['IMDb'].mean()
classif_india = df_india['IMDb'].mean()
classif_nigeria = df_nigeria['IMDb'].mean()

plt.bar(brasil, classif_brasil)
plt.bar(eua, classif_eua)
plt.bar(india, classif_india)
plt.bar(nigeria, classif_nigeria)


plt.ylabel("IMDb", size=16)
plt.xlabel("Pa√≠ses", size=16)
plt.title('M√©dia geral da classifica√ß√£o dos filmes por pa√≠ses', size=12)
plt.show()
```

```python
#filtrar anos: eua, india, brasil e nigeria

df_comedy = df.loc[(df['Genres'] == 'Comedy')]

df_documentary = df.loc[(df['Genres'] == 'Documentary')]

df_drama = df.loc[(df['Genres'] == 'Drama')]

df_romance = df.loc[(df['Genres'] == 'Romance')]
```

```python
plt.figure(figsize=(8, 6)) 
years = df.Year.unique()

classif_comedy = df_comedy.groupby('Year')['IMDb'].mean()
classif_documentary = df_documentary.groupby('Year')['IMDb'].mean()
classif_drama = df_drama.groupby('Year')['IMDb'].mean()
classif_romance = df_romance.groupby('Year')['IMDb'].mean()

ax = sns.lineplot(x=years, y=classif_comedy, label='Com√©dia')
ax = sns.lineplot(x=years, y=classif_documentary, label='Document√°rio')
ax = sns.lineplot(x=years, y=classif_drama, label='Drama')
ax = sns.lineplot(x=years, y=classif_romance, label='Romance')

ax = sns.scatterplot(x=years, y=classif_comedy, color='blue' )
ax = sns.scatterplot(x=years, y=classif_documentary, color='orange')
ax = sns.scatterplot(x=years, y=classif_drama, color='green')
ax = sns.scatterplot(x=years, y=classif_romance, color='red')

plt.legend(title='G√™neros', prop={'size': 14},  title_fontsize = 16)
plt.ylabel("IMDb", size=19)
plt.xlabel("Anos", size=19)
plt.xticks([2018, 2019, 2020, 2021], size=16)
plt.yticks([5.0, 5.5, 6.0, 6.5, 7.0, 7.5], size=16)
plt.show()
```

```python
df_two = pd.read_csv("./csv/dadosNetflix.csv")

#excluindo coluna language
df_two = df_two.drop(columns=['Language'])


df_two = df_two.dropna(how= 'any', axis = 0)
```

```python
df_ageMenos18 = df_two.loc[(df_two['Age'] == -18.0)]

df_ageMais18 = df_two.loc[(df_two['Age'] == 18.0)]
```

```python
plt.figure(figsize=(8, 6)) 
years = df.Year.unique()

classif_menos18 = df_ageMenos18.groupby('Year')['IMDb'].mean()
classif_mais18 = df_ageMais18.groupby('Year')['IMDb'].mean()

ax = sns.lineplot(x=years, y=classif_menos18, label='Menor de 18 anos')
ax = sns.lineplot(x=years, y=classif_mais18, label='Maior de 18 anos')

ax = sns.scatterplot(x=years, y=classif_menos18, color='blue')
ax = sns.scatterplot(x=years, y=classif_mais18, color='orange')

plt.legend(title='Faixa et√°ria', prop={'size': 14},  title_fontsize = 16)
plt.ylabel("IMDb", size=19)
plt.xlabel("Anos", size=19)
plt.yticks([5.9, 6.0, 6.1, 6.2, 6.3, 6.4, 6.5, 6.6], size=16)
plt.xticks([2018, 2019, 2020, 2021], size=16)

plt.show()
```
